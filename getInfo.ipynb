```python
from tqdm import tqdm
import time
def getAllInfo(rooturl, pathlist):
    allSiteInfos = {
        "data":[]
    }
    for path in tqdm(pathlist):
        _url = rooturl+path
        res = requests.get(_url)

        soup = BeautifulSoup(res.text, "html.parser")

        soup
        
        try:
            title = soup.select("title")[0].text
        except IndexError:
            title = "取得できませんでした"

        try:
            captionText = soup.select("meta[name='description']")[0]["content"]
        except IndexError:
            captionText = "取得できませんでした"

        try:
            captionImage = soup.select('meta[property="og:image"]')[0]["content"]
        except IndexError:
            captionText = "取得できませんでした"

        _dict = {
            "title": title, 
            "url": _url,
            "captionText": captionText,
            "captionImage": captionImage
        }
        allSiteInfos["data"].append(_dict)
        time.sleep(0.5)

    return allSiteInfos
```

```python
allSiteInfos = getAllInfo(rooturl, pathlist)
allSiteInfos
```

```python
import json
with open('allSiteInfos.json', 'w') as f:
    json.dump(allSiteInfos, f, indent=4, ensure_ascii=False)
```
